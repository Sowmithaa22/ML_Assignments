### Apriori algorithm

```python
import pandas as pd 
# Loading the Data without considering the first row as column headings
data = pd.read_csv(r'R:\\datasets\\store_data.csv', header=None)

# Preprocessing the data to convert it into a list of transactions
transactions = []
for index, row in data.iterrows():
    transaction = [item.strip() for item in row if isinstance(item, str)]
    transactions.append(transaction)

df=pd.DataFrame(transactions)
df['Items']=transactions
df2=pd.DataFrame()
df2['Transactions']=range(0,7501)
df2['Items']=df['Items']

# Define a simple csv reading function
def read_csv(file_path):
    data = []
    with open(file_path, 'r') as file:
        for line in file:
            data.append(line.strip().split(','))
    return data

# Define a function to preprocess the data
def preprocess_data(data):
    transactions = []
    for row in data:
        transaction = [item.strip() for item in row if isinstance(item, str)]
        transactions.append(transaction)
    return transactions

# Define a function to calculate support count
def support_count(data, itemset):
    count = 0
    for transaction in data:
        if all(item in transaction for item in itemset):
            count += 1
    return count

# Define a function to generate candidate itemsets
def generate_candidates(prev_itemsets, k):
    candidates = []
    n = len(prev_itemsets)
    for i in range(n):
        for j in range(i + 1, n):
            itemset1 = prev_itemsets[i]
            itemset2 = prev_itemsets[j]
            if itemset1[:-1] == itemset2[:-1]:
                candidate = sorted(itemset1 + [itemset2[-1]])
                if all(candidate[:i] + candidate[i + 1:] in prev_itemsets for i in range(k)):
                    candidates.append(candidate)
    return candidates

# Define a function to apply the Apriori algorithm
def apriori(data, min_support):
    itemsets = [[[item]] for item in sorted(set(item for transaction in data for item in transaction))]
    frequent_itemsets = []
    while itemsets:
        candidates = generate_candidates([itemset for sublist in itemsets for itemset in sublist], len(itemsets[0][0]) + 1)
        frequent_itemsets_level = []
        for candidate in candidates:
            support = support_count(data, candidate)
            if support >= min_support * len(data):
                frequent_itemsets_level.append(candidate)
        if frequent_itemsets_level:
            frequent_itemsets.extend(frequent_itemsets_level)
            itemsets = [frequent_itemsets_level]
        else:
            break
    return frequent_itemsets

file_path = r'R:\\datasets\\store_data.csv'
data = read_csv(file_path)
transactions = preprocess_data(data)

min_support = 0.05
frequent_itemsets = apriori(transactions, min_support)
print("Frequent Itemsets of size-2:")
for itemset in frequent_itemsets:
    print(itemset)
```

### FP-Growth algorithm

```python
import pandas as pd
from collections import defaultdict

class FPNode:
    def __init__(self, item, count, parent):
        self.item = item
        self.count = count
        self.parent = parent
        self.children = {}
        self.next = None  # Added next attribute for linked list

class FPTree:
    def __init__(self, min_support):
        self.min_support = min_support
        self.frequent_2_itemsets = []
        self.header_table = defaultdict(list)  # Changed to defaultdict(list) for linked list

    def construct_tree(self, transactions):
        self.root = FPNode(None, 0, None)
        for transaction in transactions:
            self.insert_transaction(transaction, self.root)

    def insert_transaction(self, transaction, node):
        if not transaction:
            return
        item = transaction[0]
        if item in node.children:
            node.children[item].count += 1
        else:
            node.children[item] = FPNode(item, 1, node)
            self.header_table[item].append(node.children[item])  # Added node to linked list
        self.insert_transaction(transaction[1:], node.children[item])

    def find_frequent_2_itemsets(self):
        for item in self.header_table:
            itemset = [item]
            support = 0
            for node in self.header_table[item]:
                support += node.count
            if support >= self.min_support:
                self.frequent_2_itemsets.append((itemset, support))

    def find_maximal_frequent_2_itemset(self):
        self.find_frequent_2_itemsets()
        maximal_itemset = max(self.frequent_2_itemsets, key=lambda x: x[1])
        print("Maximal Frequent Itemset of Size 2:")
        print(maximal_itemset)

# Load the dataset
data = pd.read_csv("R:\\datasets\\store_data.csv", header=None)

# Convert the dataset into a list of transactions
transactions = data.apply(lambda row: row.dropna().tolist(), axis=1).tolist()

# Set minimum support threshold
min_support = len(transactions) * 0.05

# Run FP-Growth algorithm
fp_tree = FPTree(min_support)
fp_tree.construct_tree(transactions)

# Find maximal frequent itemset of size 2
fp_tree.find_maximal_frequent_2_itemset()
```
